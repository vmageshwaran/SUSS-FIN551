```python
# Question 1
# Final Version 1.0

def calculate_quotient_and_reminder(y, d):                        # Function to accept args - Divident List, Divisor
    quoList = []                                                  # Declare Quotient List
    remList = []                                                  # Declare Reminder List
    if (d == 0) :                                                 # Handle condition where Divisor is Zero
        print ('<Error> : Invalid Input - Divisor !!!')
    else:
        for item in y:
            if (isinstance(item, int) == True):
                quoList.append(int(item/d))                        # Append to Quotient list with integer of Divident / Divisor 
                remList.append(item%d)                             # Append tp reminder list with mod of Divident / Divisor
            else:
                quoList.append('NaN')                              # Append 'NaN' to Quotient list, where divident is non int
                remList.append('NaN')                              # Append 'NaN' to Remidner list, where divident is non int
    return(quoList, remList)                                       # Return Quotient and reminder List

# Main Program
 
dividend_data = [ [20,   7,  88,   66,  18],                        # Test Data set
                  [73,  57,  67,   61,  83],
                  [14,  42,  45,   35,  33],
                  [77,  31,  44,   12,  11],
                  [22,  80,  80,   16,  84] ]

divisor_data = [9, 9, 4, 8, 4]

for ind, dividend_item in enumerate(dividend_data): 
    print('Test Case : ', ind+1)                                                    # Test Case output
    quo, rem = calculate_quotient_and_reminder(dividend_item, divisor_data[ind])    #  Function call - Divident, Divisor
    print('+'*35)                                                               
    print('Dividend : ', dividend_item)
    print('Divisor  : ', divisor_data[ind])
    print('Quotient : ', quo)
    print('Reminder : ', rem, '\n')
print('='*35)
```

```python
# Question 3
# Final Version 1.0

import math

def calculate_mortgage_cashflows(princ_amt, intrs_per, numbr_yrs, paymt_frq):  # Fn args - Principal Amount, Interest Rate, 
                                                                               # Number of years, Payment Frequency
        
    numbr_pyt = float (numbr_yrs * paymt_frq)                                  # Calculate Number of Payments 
    intrs_rat = ( (float(intrs_per) / 100) / paymt_frq )                       # Calculate Interest Rate 
    dscnt_fac = ( ((math.pow((1+intrs_rat), numbr_pyt ) -1)) / \
                  ( intrs_rat *((math.pow((1+intrs_rat), numbr_pyt))) ) )      # Calculate Discount Factor
    paymt_amt = (princ_amt / dscnt_fac)                                        # Calculate Payment Amount per frequency
    total_paymt_amt = (paymt_amt * numbr_pyt )                                 # Calculate Total Payment Amount
    total_intrs_amt = ( total_paymt_amt - princ_amt)                           # Calculate Total Interest Amount
    return paymt_amt, total_paymt_amt, total_intrs_amt              # Return, 

# Main Program
 
testcase_data = [ # ['$1,000,000', 10,    'yearly', '1.5%'],                        # Test Data set
                    ['$1,131,000', 15, 'quarterly',  '29%'], 
                    ['$122,000',   21,   'monthly',   '8%'],  
                    ['$331,000',   23,    'yearly',  '19%'],  
                    ['$1,397,000', 24,   'monthly',  '11%'],  
                    ['$125,000',    3, 'quarterly',  '24%']  
                 ]

for item_set_ind, item_set in enumerate(testcase_data): 
    my_princ_amt = float(item_set[0].replace('$','').replace(',',''))          # Convert input to number float format
    my_numbr_yrs = float(item_set[1])                                          # Convert input to number float format
    my_paymt_frq = float((item_set[2] \
                          .replace('quarterly',  '4') \
                          .replace(  'monthly', '12') \
                          .replace(   'yearly',  '1'))  )                      # Convert frequency text to number float format
    my_intrs_rat = float(item_set[3].replace('%',''))                          # Convert input to number float format
    my_paymt_amt, my_total_paymt_amt, my_total_intrs_amt = \
                   calculate_mortgage_cashflows \
                      (my_princ_amt, my_intrs_rat, my_numbr_yrs, my_paymt_frq) #  Function call for loan calculation
    
    print('Test Case : ', item_set_ind + 1)                                    # Test Case output to the nearest dollar
    print('+'*45)                                                               
    print('Principal Amount      : ${:,.0f}'.format(my_princ_amt))
    print('Interest Rate         : {}'.format(item_set[3]))
    print('Number of Years       : {:,.0f}'.format(my_numbr_yrs))
    print('Payment Frequency     : {}'.format(item_set[2]))
    #print('Payment Frequency     : {} : {}'.format(item_set[2], my_paymt_frq))
    print('Payment Amount        : ${:,.0f}'.format(my_paymt_amt))
    print('Total Payment Amount  : ${:,.0f}'.format(my_total_paymt_amt))
    print('Total Interest Amount : ${:,.0f}'.format(my_total_intrs_amt), '\n')
print('='*45)                     
  
```

```python
# Question 4
#Fincal Version  1.0

import numpy  as np
import pandas as pd

MISSING_VALUE = -99                                                                # Definiton of Missing value indicator 

def data_cleaner(inp_ds):                                                          # Function with argument as list
    unclean_ds = inp_ds
    col_list = [ 'shift_up', 'shift_down', 'cleaned',  'replaced']                 # Column list definition
    
    dsDF = pd.DataFrame(unclean_ds, columns=['unclean'])                           # df using the input dataset
    dsDF = dsDF.reindex(dsDF.columns.tolist() + col_list, axis=1)                  # df additional columnn definition

    dsDF['cleaned']    = dsDF['unclean'].replace(MISSING_VALUE, np.NaN)            # df clean-up step to replace NaN
    dsDF['shift_up']   = dsDF['unclean'].shift(-1).replace(MISSING_VALUE, np.NaN)  # df clean-up step to shift up
    dsDF['shift_down'] = dsDF['unclean'].shift(1).replace(MISSING_VALUE, np.NaN)   # df clean-up step to shift down
    dsDF['replaced']   = dsDF[dsDF['unclean'] != MISSING_VALUE]
    dsDF['replaced']   = dsDF['replaced'].fillna(dsDF[['shift_up', 'shift_down']].mean(axis=1)) # df step to derive adj mean
    return dsDF 


# Main Program
 
testcase_data = [ [1,    1,  -99,   1,   1, -99,   1, -99],                        # Test Data set
                  [2,    2,    2,   2,   2, -99,   2, -99],
                  [3,  -99,    3,   3, -99,   3,   3, -99],
                  [4,    4,    4,   4, -99,   4, -99, -99],
                  [5,    5,    5, -99,   5,   5, -99, -99] ]
                
cycle_cnt = 1                                                                      # Iteration cycel count
for item_set in testcase_data:                                                     # Loop thru the testcase dataset
    #print('item_set : ', item_set, type(item_set))
    myDF_Finish = True                                                             # Flag to handle interative calls
    myDF_Firstrun = True                                                           # First run indicator
    while myDF_Finish == True :                                                    # while loop till the flag remains true
          myDF = data_cleaner(item_set)                                            # function call
        
          if myDF_Firstrun == True :
             unclean_set = myDF['unclean'].to_list()                               # preserve the unclean_set 
             myDF_Firstrun = False
            
          if myDF['replaced'].isnull().any() :
             item_set = myDF['replaced'].to_list()                                 # preparation step for interative call
          else :
             myDF_Finish = False
                
    clean_set = myDF['replaced'].astype(int).to_list()                             # Cleaned dataset
     
    print('Test Case : ', cycle_cnt)                                               # Test case output
    print('+'*60) 
    print('Original Dataset : ', unclean_set)
    print('Cleaned  Dataset : ', clean_set)
    if (unclean_set != myDF['replaced'].to_list()):                               # Replacement message
       print ('!!! Data clean-up / replacement of missing data DONE !!!!','\n')
    else:
       print ('*** No Data clean up done ***', '\n')
    cycle_cnt = cycle_cnt + 1                                                     # increment cycle count
print('='*60)
```
